If else statements in python

```python
#to check email and pass are correct or not 
#1st way to write used nested elif
email = input("enter ur email")
password = input("enter password")

if email == 'avjain@gmail.com' and password == '1234':
    print("welcome")
elif email == 'avjain@gmail.com' and password != '1234': 
    print("try again wrong pass")
    password = input("enter password again")
    if password == '1234':
        print("good...finally ")
    else:
        print("not appplicable ")
else:
    print("not valid")

#another way to write is this nested if
if email == 'avjain@gmail.com':
    print("yes now go with pass")
    if password == '234':
        print("you are in")
    else:
        print("hard luck")
else:
    print("not valid...byee")
```

```python
#if else 
# find max of 3 numbers
a = int(input("enter ur first num: "))
b = int(input("enter ur second num: "))
c = int(input("enter ur third num: "))
print("Comparing the numbers")

if a >= b and a >= c:
    print(a,"is max ")
elif b>=c and b>=a:
    print(b,"is max ")
else:
    print(c,"is max")
```

```python
#menu driven program 
fnum = int(input("enter first num"))
snum = int(input("enter second num"))

op = input("enter the operator u wish to use: ")

if op == '+':
    print("add",fnum + snum)
elif op == '-':
    print("sub",fnum - snum)
elif op == '*':
    print("multi",fnum * snum)
else:
    print("divide",fnum/snum)
```
