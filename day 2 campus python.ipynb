```python
#there are differen types of operators 
#arithmetic 
print(5+6)
print(5-6)
print(5/6)
print(5%6)
print(5//6)
print(5*6)
print(5**2)
```

```python
#relational operators 
```

```python
print(4>5)
print(4<5)
print(4>=5)
print(5<=6)
print(4==4)
print(4!=4)
```

```python
#logical operators
print(1 and 0)
print(not 1)
print(1 or 0)
```

```python
#bitwise operators 
print(2 & 3)
print(2 | 3)
print(2 ^ 3)
print( ~3)
print(4 >> 2)
print(5 << 2)
```

```python
#assignment operators 
a = 10 
# a = a/2
a/=2
print(a)
```

```python
#membership operator 
#in and not in
print('D' in 'DELHI')
print('D' not in 'DELHI')
#also available in different data types 
print(1 in [2,3,4,5])
```
